```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
data=pd.read_csv("data.csv")
data
```

```python
type(data)
```

# Fiiling age here

```python
imputer=SimpleImputer( missing_values=np.nan,strategy='mean')
imputer.fit(data.iloc[:,1:2].values)

data.iloc[:,1:2]=imputer.transform(data.iloc[:,1:2].values)
```

```python
data
```

# Filling salary here

```python
imputer=SimpleImputer( missing_values=np.nan,strategy='mean')
imputer.fit(data.iloc[:,2:3].values)

data.iloc[:,2:3]=imputer.transform(data.iloc[:,2:3].values)
```

```python
data
```

# Filling all values using most frequent

```python
imputer=SimpleImputer( missing_values=np.nan,strategy='most_frequent')
imputer.fit(data.iloc[:,:].values)

data.iloc[:,:]=imputer.transform(data.iloc[:,:].values)
```

```python
data
```

```python
data.isnull()
```

```python
data.isnull().sum()
```
